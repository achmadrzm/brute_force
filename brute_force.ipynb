```python
import itertools
import string

chars = string.ascii_lowercase + string.ascii_uppercase + string.digits + '`' + '~' + '!' + '@' + '#' + '$' + '%' + '^' + '&' + '*' + '(' + ')' + '-' + '_' + '=' + '+' + '[' + '{' + ']' + '}' + '|' + ';' + ':' + ',' + '<' + '.' + '>' + '/' + '?'

coba = 0
mulai = 1
selesai = 10

for panjang_password in range (mulai, selesai):
  for tebak in itertools.product (chars, repeat = panjang_password):
    coba += 1
    tebak = ''.join (tebak)
    print (tebak,coba)
```
